```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

new_data['Private'] = le.fit_transform(new_data['Private'])
new_data['PhD'] = le.fit_transform(new_data['PhD'])
```

```python
from scipy.stats import pearsonr

stat, p = pearsonr(new_data['Private'], new_data['PhD'])
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
 print('Probably independent')
else:
 print('Probably dependent')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
# Create Linear Regression Object
lin_reg = LinearRegression()
X1 = new_data[['Private', 'Apps', 'Accept', 'Enroll', 'Students', 'S.F.Ratio', 'Expend', 'PhD']]
Y1 = new_data[['Grad.Rate']] 
x_train, x_test, y_train, y_test = train_test_split(X1, Y1, test_size=0.20, random_state=45)
# Fit (Train) the model
lin_reg.fit(x_train,y_train)
print("Intercept for the linear regresison model is", lin_reg.intercept_, "and the scope is",lin_reg.coef_)
print("\n")
# Prediction using Training Data
train1 = lin_reg.predict(x_train)
print(train1[0:5])
# Prediction using Test Data
test1 = lin_reg.predict(x_test)
print(test1[0:5])
print("\n")
```

```python
# Model accuracy using Test Data
mse_test = mean_squared_error(y_test,test1)
print("Mean square error is",mse_test)
print("R-Square value of test data is", lin_reg.score(x_test,y_test))
print("\n")
# Model accuracy using Training Data
mse_train = mean_squared_error(y_train,train1)
print("Mean square error is",mse_train)
print("R-Square value of training data is", lin_reg.score(x_train,y_train))
```

```python
from sklearn.model_selection import cross_val_score, KFold

cross = cross_val_score(lin_reg, X1, Y1, cv=KFold(n_splits=5,shuffle=True))

print(cross)
print("The mean of the folds are", cross.mean())
```
